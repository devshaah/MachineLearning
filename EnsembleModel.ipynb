```python
 #we train multiple models on same dataset
 #bagging and boosting are major ensemble techniques
 #we train models on subset of dataset , ie model with smaller dataset and are called weak learners
 #when such weak learners are combined , we overall get a good result
 #bagging is also called bootstrap-aggregation , eg of bagging is RandomForest
 #bootstrap -> creating subsets of data from dataset
 #aggregation -> considering all the weak learners to get a good output
```

```python
import pandas as pd

df = pd.read_csv('diabetes.csv')
df.isnull().sum() #checking for null values
```

```python
df.describe() #statistics of all columns
```

```python
X = df.drop(['Outcome'],axis='columns')
y = df['Outcome']

from sklearn.preprocessing import StandardScaler
x_scaled = StandardScaler().fit_transform(X)

from sklearn.model_selection import train_test_split
xt,xtst ,yt,ytst = train_test_split(x_scaled,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

ans = cross_val_score(DecisionTreeClassifier(),xt,yt,cv=5)
ans.mean()
```

```python
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    base_estimator=DecisionTreeClassifier(),
    n_estimators=100, #no of datasets
    max_samples = 0.8 , #size of subset
    oob_score = True #possibility of missing something , so we use that as test
)
bag_model.fit(xt,yt)
bag_model.score(xtst,ytst)

#using crossvalscore
bag_scores = cross_val_score(bag_model,xt,yt,cv=5)
bag_scores.mean()
```
