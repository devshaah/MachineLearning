```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
df = pd.read_csv('insurance_data.csv')
# plt.scatter(df['age'],df['bought_insurance'],color='red')

x_trn, x_tst, y_trn, y_tst = train_test_split(df[['age']],df['bought_insurance'],test_size=0.1)

model = linear_model.LogisticRegression()

model.fit(x_trn,y_trn)
# plt.plot(x_trn,model.predict(x_trn),color='blue')

model.predict(x_tst)
model.score(x_tst,y_tst)
```

```python
# SUPPORT VECTOR MACHINE
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.svm import SVC
df = pd.DataFrame(load_iris().data, columns = load_iris().feature_names)
df['target'] = load_iris().target
df['name'] = df.target.apply(lambda x : load_iris().target_names[x])

df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]


plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'] , color = 'blue')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'] , color = 'red')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'] , color = 'yellow')

X_train, X_test, y_train, y_test = train_test_split(df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']], df['target'] ,test_size=0.2)
model = SVC()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
