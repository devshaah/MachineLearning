```python
# PROCESS OF FIGUIRNG OUT MOST IMPORTANT FEATURES THAT HAS THE MOST IMPACT ON TARGET VARIABLE
# PCA (n_components = 6 ) figures out the most 6 important features that have the impact
# this helps in gtraining the model faster
# few key points :  scale before applying pca , accuracy may reduce
# also called dimension rductionality technique as it helps us in redcing the dimensions
```

```python
import pandas as pd
from sklearn.datasets import load_digits

digits = load_digits()
df = pd.DataFrame(digits.data , columns=digits.feature_names)
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(df)
x_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_scaled,digits.target, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95) #retain 95% of useful features
x_pca = pca.fit_transform(df)
x_pca.shape #from 64 columns,we are left only with 29 useful features

X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(x_pca,digits.target, test_size=0.2)
model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca,y_train_pca)
model.score(X_test_pca,y_test_pca)

#EVEN AFTER DROPPING FEATURES, THE ACCURACY REMAINED ALMOST SAME
```

```python
# EXERCISE
df = pd.read_csv('heart.csv')

df['Sex'] = df['Sex'].apply(lambda x :1 if x=='M' else 0)
df['ExerciseAngina'] = df['ExerciseAngina'].apply(lambda x :1 if x=='Y' else 0)

df.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df.ChestPainType.replace(
    {
        'ATA':1, 'NAP':2, 'ASY':3, 'TA':4
    },
    inplace=True)

X =df.drop(['HeartDisease'],axis='columns')
Y = df['HeartDisease']

scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)

pca = PCA(0.95)
x_pca = pca.fit_transform(x_scaled)

X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(x_pca,Y, test_size=0.2)
model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca,y_train_pca)
model.score(X_test_pca,y_test_pca)
```
